```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

# Работа с тренировочными данными.

Загрузим тестовые данные и поменяем имена колонок.

```python
df_train = pd.read_csv(train_path, encoding='cp1251')
columns = [
    'id', 'id_bulk', 'spalen', 'date1', 'start_square', 'value', 'price', 
    'mean_sq', 'mean_fl', 'plan_s', 'plan_m', 'plan_l', 'vid_0', 'vid_1', 
    'vid_2', 'month', 'month_cnt', 'class_objecta', 'kolichestvo_pomecheniy', 
    'ogorojena_territoriya', 'plochad_zemelnogo_uchastka', 'vhodnie_gruppi',
    'detskiy_sad', 'school', 'hospital', 'fok', 'sportivnaya_plochadka',
    'avtomoika', 'kladovie', 'kolyasochnie', 'kondei', 'ventilaciya', 'lift',
    'sistema_musora_otvedeniya', 'videonabludenie', 'podzemnaiya_parkovka',
    'dvor_bez_machin', 'machinomesta', 'plochad_promzoni_radius_500_m',
    'plochad_zelenoy_zoni_radius_500_m', 'do_kremliya_km', 'do_ttk_km',
    'do_sadovogo_km', 'do_bolchoy_dorogi_na_machine',
    'do_udobnoy_avtorazviyazki_km', 'do_metro_pechkom_km', 'do_promki_km',
    'do_parka_km', 'do_parka_pechkom_km','stanciy_metro_ot_kolca',
    'plochad_dvora', 'kurs', 'stavka_po_ipoteke', 'vkladi_ot_1_goda',
    'vkladi_ot_1do_3_let', 'vkladi_sviche_3_let'
]
df_train.columns = columns
```

```python
df_train.head()
```

Уберем столбцы, которых нет в тестовых данных.

```python
df_train.drop(['start_square', 'plan_s', 'plan_m', 'plan_l', 'vid_0', 'vid_1', 'vid_2'], 
        axis=1, inplace=True)
```

Сделаем one-hot encoding для class_objecta и для sistema_musora_otvedeniya.

```python
size = df_train.shape[0]
l = pd.Series([0] * size)
df_train['econom'] = l
df_train['comfort'] = l
df_train['standart'] = l
```

```python
for index, row in df_train.iterrows():
    class_obj = row['class_objecta']
    if class_obj == 'комфорт':
        df_train.loc[index, 'comfort'] = 1
    elif class_obj == 'эконом':
        df_train.loc[index, 'econom'] = 1
    else:
        df_train.loc[index, 'standart'] = 1
        
df_train['sistema_musora_otvedeniya'] = 1
```

Заменим "да" и "нет" на 1 и 0 соответсвенно в столбцах ogorojena_territoriya, avtomoika, kladovie, kolyasochnie, podzemnaiya_parkovka, dvor_bez_machin, sportivnaya_plochadka и vhodnie_gruppi.

Удалим лишние столбцы.

```python
df_train.drop('class_objecta', axis=1, inplace=True)
df_train.drop('date1', axis=1, inplace=True)
df_train.drop('id_bulk', axis=1, inplace=True)
df_train.drop('id', axis=1, inplace=True)
```

Выберем все столбцы кроме value.

```python
indexes = [i for i in range(df_train.columns.shape[0])]
chosen_columns = df_train.columns[indexes[0:1] + indexes[2:]]
```

Проверим, что все признаки получились числовыми. Для этого возьмем первый 0-ую строку без ограничения общности.

```python
df_train[chosen_columns].loc[0, :]
```

Посмотрим, как признаки коррелируют между собой и выберем те, у которых она большая.

# Займемся препроцессингом тренировочных данных.

```python
y_train = df_train['value'].values
X_train = df_train.drop('value', axis=1)
```

# Построение моделей.

Запишем функцию для вычисления RMSE

```python
def compute_rmse(y, pred):
    result = np.sqrt(np.sum((y - pred) ** 2)) / np.sqrt(y.shape[0])
    return result
```

Посмотрим, как ведут себя стандартные алгоритмы.

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

train_scores = list()
test_scores = list()

# alphas = [1, 2, 3, 4, 5]
alphas = [10, 20, 30, 50, 100, 250]
for alpha in alphas:
    tr_scores = list()
    t_scores = list()
    
    for train_index, test_index in kf.split(X_train):
        X_train_kf = X_train[train_index]
        y_train_kf = y_train[train_index]
        X_test_kf = X_train[test_index]
        y_test_kf = y_train[test_index]
        
        # model = Ridge(alpha=alpha)
        # model = Lasso(alpha=alpha)
        # model = RandomForestRegressor(n_estimators=alpha, random_state=42)
        # model = KNeighborsRegressor(n_neighbors=alpha)
        model = GradientBoostingRegressor(n_estimators=alpha, random_state=42)
        model.fit(X_train_kf, y_train_kf)
        
        tr_scores.append(compute_rmse(model.predict(X_train_kf), y_train_kf))
        t_scores.append(compute_rmse(model.predict(X_test_kf), y_test_kf))
        
    train_scores.append(np.array(tr_scores).mean())
    test_scores.append(np.array(t_scores).mean())
```

```python
plt.plot(alphas, train_scores, label='train')
plt.plot(alphas, test_scores, label='test')
plt.grid()
plt.legend()
```
